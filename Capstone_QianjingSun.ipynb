```python
# Install the required lib
!pip install scikit-learn pandas matplotlib --quiet

# 1. Import the lib
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Import the data from google drvie
df = pd.read_csv("/content/drive/My Drive/Capstone Project2025/phishing_email.csv")  # make sure the file path is right
print("Columns:", df.columns.tolist())
print(df[['text_combined', 'label']].head(7))

```

```python
# 3. Read the input and label
X = df['text_combined'].fillna('')  # 避免空值
y = df['label']  # 0: ham, 1: spam/phishing


```

```python
# 4. TF-IDF 文本向量化
vectorizer = TfidfVectorizer(stop_words='english', max_features=5500)
X_vec = vectorizer.fit_transform(X)

```

```python
# 5. 拆分训练集与测试集
X_train, X_test, y_train, y_test = train_test_split(X_vec, y, test_size=0.2, random_state=42)
```

```python

# 6. 训练 KNN 模型（K=3，可调）
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

```

```python
# 7. 预测并评估
y_pred = knn.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

```

```python
# 8. 混淆矩阵可视化
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Ham", "Spam"], yticklabels=["Ham", "Spam"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```
